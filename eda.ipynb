```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
pd.set_option('display.width',500)
df=sns.load_dataset("titanic")
df.head()
```

```python
def check_df(dataframe,head=5):
    print("####### Shape #######")
    print(dataframe.shape)
    print("####### Types #######")
    print(dataframe.dtypes)
    print("####### Head #######")
    print(dataframe.head(head))
    print("####### Tail #######")
    print(dataframe.tail(head))
    print("####### NA #######")
    print(dataframe.isnull().sum())
    print("####### Quantiles #######")
    print(dataframe.describe([0,0.05,0.50,0.95,0.99,1]).T)

check_df(df)
```

## Analysis of Categorical Variables


```python
#Categorical values
cat_cols=[col for col in df.columns if str(df[col].dtypes) in ["category","object","bool"]]
print(cat_cols)
```

```python
#how can we find categorical values which has numerical dtype? 

num_but_cat=[col for col in df.columns if df[col].nunique()<10 and str(df[col].dtypes) in ["int64","float64"]]
print(num_but_cat)

#so in real and all categorical values is:
cat_cols=cat_cols+num_but_cat
print(cat_cols)
```

```python
#getting a summary of categorical values

def cat_summary(dataframe,col_name,plot=False):
    print(pd.DataFrame({col_name: dataframe[col_name].value_counts(),
                        "Ratio": 100*dataframe[col_name].value_counts()/len(dataframe)}))
    print("############")
    if plot:
        sns.countplot(x=dataframe[col_name],data=dataframe)
    plt.show(block=True)
cat_summary(df,"sex",plot=True)
```

```python
#Loop for playing all categorical variables

for col in cat_cols:
    if df[col].dtypes=="bool":
        df[col]=df[col].astype(int)
    else:
        cat_summary(df,col,plot=True)
```

```python
# Getting summary for all this codes:
def cat_summary(dataframe,col_name,plot=False):
    if dataframe[col_name].dtype=="bool":
        dataframe[col_name]= dataframe[col_name].astype(int)
        if plot:
            sns.countplot(x=dataframe[col_name],data=dataframe)
            plt.show(block=True)
    else:
        print(pd.DataFrame({col_name: dataframe[col_name].value_counts(),
                            "Ratio": 100*dataframe[col_name].value_counts()/len(dataframe)}))
        print("############")
        if plot:
            sns.countplot(x=dataframe[col_name],data=dataframe)
        plt.show(block=True)

cat_summary(df,"adult_male",plot=True)
```

## Analysis of Numerical Variables


```python
num_cols=[col for col in df.columns if df[col].dtypes in ["int64","float64"]]
num_cols=[col for col in num_cols if col not in cat_cols]
print(num_cols)
```

```python
def num_summary(dataframe,numerical_col,plot=False):
    quantiles=[0.05,0.10,0.20,0.30,0.40,0.50,0.60,0.70,0.80,0.90,0.95,0.99]
    print(dataframe[numerical_col].describe(quantiles).T)
    if plot:
        dataframe[numerical_col].hist()
        plt.xlabel(numerical_col)
        plt.title(numerical_col)
        plt.show()

for col in num_cols:
    num_summary(df,col,plot=True)
```

## Capturing Variables and Generalizing Operations

```python
def grab_col_names(dataframe,cat_th=10,car_th=20):
    cat_cols=[col for col in df.columns if str(df[col].dtypes) in ["category","object","bool"]]
    num_but_cat=[col for col in df.columns if df[col].nunique()<10 and df[col].dtypes in ["int64","float64"]]
    cat_but_car=[col for col in df.columns if df[col].nunique()>20 and str(df[col].dtypes) in ["category","object"]]
    cat_cols=cat_cols+num_but_cat
    cat_cols=[col for col in cat_cols if col not in cat_but_car]

    num_cols=[col for col in df.columns if df[col].dtypes in ["int64","float64"]]
    num_cols=[col for col in num_cols if col not in cat_cols]

    print(f"Observation: {dataframe.shape[0]}")
    print(f"Variables:{dataframe.shape[1]}")
    print(f"cat_cols:{len(cat_cols)}")
    print(f"num_cols:{len(num_cols)}")
    print(f"cat_but_car:{len(cat_but_car)}")
    print(f"num_but_cat:{len(num_but_cat)}")

    return cat_cols,num_cols,cat_but_car

cat_cols,num_cols,cat_but_car=grab_col_names(df)
```

## Analysis of Correlation


```python
df=pd.read_csv("data.csv")
df=df.iloc[:,1:-1]
df.head()
```

```python
num_cols=[col for col in df.columns if df[col].dtype in [int,float]]
corr=df[num_cols].corr()
corr
```

```python
sns.set(rc={"figure.figsize":(12,12)})
sns.heatmap(corr,cmap="RdBu")
plt.show();
```

```python
cor_matrix=df.corr().abs()
```

```python
upper_triangle_matrix=cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(np.bool))
upper_triangle_matrix
```

```python
drop_list=[col for col in upper_triangle_matrix if any(upper_triangle_matrix[col]>0.90)]
drop_list
```

```python
df.drop(drop_list,axis=1)
```

```python
# Doing these process through function:

def high_correlated_cols(dataframe,plot=False,corr_th=0.90):
    corr=dataframe.corr()
    corr_matrix=corr.abs()
    upper_triangle_matrix=cor_matrix.where(np.triu(np.ones(corr_matrix.shape),k=1).astype(np.bool))
    drop_list=[col for col in upper_triangle_matrix.columns if any(upper_triangle_matrix[col]>corr_th)]
    if plot:
        sns.heatmap(corr,cmap="RdBu")
        plt.show();
    return drop_list

drop_list= high_correlated_cols(df)
```

```python
high_correlated_cols(df.drop(drop_list,axis=1),plot=True)
```
