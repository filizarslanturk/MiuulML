```python
from statistics import mode
df["deck"]=df["deck"].fillna(df["deck"].mode,inplace=True)
df["deck"].isnull().sum()
```

```python
df["age"].fillna(np.median(df["age"].dropna()),inplace=True)
df["age"].isnull().sum()
```

```python

sum_df=df.groupby(["pclass","sex"]).agg({"survived":["sum","count","mean"]})
sum_df
```

```python
def age_flag(age):
    if age<30:
        return 1
    else:
        return 0

df["age"]=df["age"].apply(lambda x: age_flag(x))
df["age"]
```
